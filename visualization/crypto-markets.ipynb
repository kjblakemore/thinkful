## Drill: Present the same data multiple ways (Unit 1/Lesson 2/Project 3)
Data set used is [Every Cryptocurrency Daily Market Price](https://www.kaggle.com/jessevent/all-crypto-currencies/data)

```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
%matplotlib inline
```

### Choose one variable and plot the variable four different ways.

```python
markets = pd.read_csv('crypto-markets.csv')
display(markets[:5])

volume = markets.volume
```

```python
markets_bitcoin = markets[markets.slug == 'bitcoin']
volume_bitcoin = markets_bitcoin.volume
```

```python
sns.distplot(volume_bitcoin)
```

```python
sns.distplot(volume_bitcoin, bins=20, kde=False)
```

```python
sns.distplot(volume_bitcoin, hist=False)
```

```python
sns.kdeplot(volume_bitcoin, shade=True, cut = 0)
```

### Choose two continuous variables, and plot them three ways

```python
bitcoin = markets[(markets.date >= '2017-01-01') & (markets.slug =='bitcoin')][['volume', 'close']]
display(bitcoin[:5])
```

```python
sns.jointplot(x='volume', y='close', data=bitcoin)
```

```python
sns.jointplot('volume', 'close', data=bitcoin, kind='reg')
```

```python
sns.jointplot('volume', 'close', data=bitcoin, kind='hex')
```

### Choose one continuous variable and one categorical variable, and plot them six different ways

```python
large_markets = markets[markets.market >= 1000000000].groupby('slug').filter(lambda x: len(x) > 100)
display(large_markets)

print(large_markets.slug.value_counts())
```

```python
plt.figure(figsize=(15,8))

sns.stripplot(x='slug', y='market', data=large_markets)
```

```python
plt.figure(figsize=(15,8))
sns.stripplot(x='slug', y='market', data=large_markets, jitter=True)
```

```python
plt.figure(figsize=(15,8))
sns.swarmplot(x='slug', y='market', data=large_markets)
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(x='slug', y='market', data=large_markets)
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x='slug', y='market', data=large_markets)
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x='slug', data=large_markets, palette='Greens_d')
```

### With a dataset containing at least four continous variables and one categorical, create a master plot that gives insight into the variables and their interrelationships, including:
* Probability distributions
* Bivarite relationships
* Whether the distributions or the relationships vary across groups

```python
sns.set(style='ticks', color_codes=True)
sns.pairplot(large_markets, size=3, vars=['open', 'close', 'high', 'low'], hue='slug')
```
