```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt

from scipy.stats import ttest_ind

%matplotlib inline
```

## Compute the p-value for binomial distributions with varying probabilities, trial sizes and sample sizes.

```python
pop1 = np.random.binomial(10, 0.2, 10000)   # trial size = 10, probability of success = .2
pop2 = np.random.binomial(10, 0.5, 10000)   # trial size = 10, probability of success = .5

# Sample size = 100
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 

plt.subplot(1, 2, 2)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print('sample 1 mean: ', sample1.mean(), '\tstd: ', sample1.std())
print('sample 2 mean: ', sample2.mean(), '\tstd: ', sample2.std())

print(ttest_ind(sample2, sample1, equal_var=False))
```

```python
# Increase sample size from 100 to 1000
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 

plt.subplot(1, 2, 2)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print('sample 1 mean: ', sample1.mean(), '\tstd: ', sample1.std())
print('sample 2 mean: ', sample2.mean(), '\tstd: ', sample2.std())

print(ttest_ind(sample2, sample1, equal_var=False))
```

```python
# Decrease sample size to 20
sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 

plt.subplot(1, 2, 2)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print('sample 1 mean: ', sample1.mean(), '\tstd: ', sample1.std())
print('sample 2 mean: ', sample2.mean(), '\tstd: ', sample2.std())

print(ttest_ind(sample2, sample1, equal_var=False))
```

### What values change and which remain the same, when the sample size is changed?

The pvalue accuracy increases as the sample size increases.

```python
# Change the probability of success values from 0.2 and 0.5 to 0.3 and 0.4.
pop1 = np.random.binomial(10, 0.3, 10000)   # trial size = 10, probability of success = .3
pop2 = np.random.binomial(10, 0.4, 10000)   # trial size = 10, probability of success = .4

# Sample size = 100
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 

plt.subplot(1, 2, 2)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print('sample 1 mean: ', sample1.mean(), '\tstd: ', sample1.std())
print('sample 2 mean: ', sample2.mean(), '\tstd: ', sample2.std())

print(ttest_ind(sample2, sample1, equal_var=False))
```

### What changes when the probability of success values change?

As the probability of success values become more similar, the pvalue increases, since the distributions become more similar.

```python
pop1 = np.random.poisson(1.0, 10000)   # expectation of interval = 1.0, sample size
pop2 = np.random.poisson(.5, 10000)

# Sample size = 100
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 

plt.subplot(1, 2, 2)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print('sample 1 mean: ', sample1.mean(), '\tstd: ', sample1.std())
print('sample 2 mean: ', sample2.mean(), '\tstd: ', sample2.std())

print(ttest_ind(sample2, sample1, equal_var=False))
```

### When the distribution type is changed to Poisson, do the sample mean values still accuractely represent the population values?

Yes, as can be seen above, the sample means are the average value for the populations.
