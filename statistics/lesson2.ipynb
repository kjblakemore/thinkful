## DRILL - Descriptive Statistics and Normality
1. Graph six probability distributions with their corresponding means and standard deviations.
2. Graph the sum of two normal distributions along with the mean and standard deviation of the summed distributions.

```python
# 1.1 Bernoulli
bernoulli = np.random.binomial(1, .5, 100)
mean = bernoulli.mean()
std = bernoulli.std()

plt.hist(bernoulli, color='c')
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2)
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# 1.2 Uniform

uniform = np.random.uniform(0, 1, 100)
mean = uniform.mean()
std = uniform.std()

plt.hist(uniform, color='c')
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2)
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# 1.3 Binomial
binomial = np.random.binomial(20, .5, 100)
mean = binomial.mean()
std = binomial.std()

plt.hist(binomial, color='c')
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2)
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# 1.4 Normal
normal = np.random.normal(0, 1, 100)
mean = normal.mean()
std = normal.std()

plt.hist(normal, color='c')
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2)
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# 1.5 Poisson
poisson = np.random.poisson(1.0, 100)
mean = poisson.mean()
std = poisson.std()

plt.hist(poisson, color='c')
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2)
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# 1.6 Exponential
exponential = np.random.exponential(1.0, 100)
mean = exponential.mean()
std = exponential.std()

plt.hist(exponential, color='c')
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2)
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2)
plt.show()
```

Evaluate whether the descriptive statistics provided useful information about the variable. Can you identify any common characteristics of the distributions that could be usefully described using the mean and/or standard deviation, versus the ones that could not?

The descriptive statistics are most useful for normally distributed data, because then, the standard deviation shows the spread of data centered on the mean.  

```python
# 2. Graph the sum of two normal distributions along with the mean and standard deviation of the summed distributions.
#    The resulting distribution is also normal with a mean equal to the sum of the original means and a standard deviation 
#    equal to the larger of the two original standard deviations.

normal1 = np.random.normal(5, .5, 10000)
normal2 = np.random.normal(10, 1.0, 10000)
normal = normal1 + normal2
mean = normal.mean()
std = normal.std()
print('mean: ', mean, 'std:', std)

plt.hist(normal, color='c')
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2)
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2)
plt.show()
```
